```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/drive/MyDrive/Data Set/covid.csv')
plt.figure(figsize=(15,5))
plt.xlabel("Country")
plt.ylabel("Confirmed/deaths")
plt.title("Covide_19_analysis")
plt.plot(df['Country'],df['Confirmed'],label='Confirmed')
plt.plot(df['Country'],df['Deaths'],label='Deaths')
pick_confirmed=max(df['Confirmed'])
pick_daths=max(df["Deaths"])
plt.annotate(xy=[df[df.Confirmed==df.Confirmed.max()].Country,pick_confirmed],s='Confirmed peak')
plt.annotate(xy=[df[df.Deaths==df.Deaths.max()].Country,pick_daths],s='Death peak')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
